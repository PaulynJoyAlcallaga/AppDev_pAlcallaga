###Exercise 2

```python
#Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

#Enter Hours: 45
#Enter Rate: 10
#Pay: 475.0

hr = int(input("Enter Hours: "))
rt = float(input("Enter Rate: "))

if hr > 40:
  rp = 40 * rt
  op = (hr - 40) * rt * 1.5
  p = rp + op
else:
     p = hr * rt

print("Pay: " , p)
```
