
## CHALLENGE


```python

def format_name(full_name):

    parts = full_name.strip().split()

    formatted_name = parts[0].capitalize()

    if len(parts) > 1:
        formatted_name += " " + parts[1].capitalize()

    if len(parts) > 2:
        formatted_name += " " + parts[-1].capitalize()

    return formatted_name


while True:
    full_name = input("Enter a full name (or type 'quit' to exit): ")

    if full_name.lower() == 'quit':
        print("Exiting program...")
        break

    formatted_name = format_name(full_name)

    print("Formatted name:", formatted_name)
```
